```python
# EEG preprocessing & epoching pipeline (MNE‑Python)
# — minimal comments, concise style —

import os
import mne
from mne.preprocessing import ICA, compute_current_source_density

processing   = ["control", "asmr"]  # "control" or "asmr"

for proc in processing:
    print(f"Processing {proc} data...")
    raw_dir      = f"./{proc}/"
    output_dir   = f"/home/s.dharia-ra/Shyamal/ASMR_NEW/statstical_analysis/{proc}_epochs_test"
    os.makedirs(output_dir, exist_ok=True)

    # list all BrainVision header files
    files = [f for f in os.listdir(raw_dir) if f.endswith(".vhdr")]
    print("Files found:", files)

    # event code remapping
    mapping       = {2: 99, 4: 88, 6: 99, 8: 88, 10: 99, 12: 88}
    new_event_ids = {"Open Eyes": 99, "Closed Eyes": 88}

    subject_num = 1
    for fname in files:
        file_path = os.path.join(raw_dir, fname)
        raw = mne.io.read_raw_brainvision(file_path, preload=True)

        # 0) channel metadata
        raw.set_channel_types({"VEOG": "eog", "HOEG": "eog"})
        raw.set_montage(
            mne.channels.make_standard_montage("standard_1020"),
            on_missing="ignore",
        )
        events, _ = mne.events_from_annotations(raw)
        # raw.set_eeg_reference("average", projection=False)
        raw.resample(250, events = events)

        # 1) ICA‑friendly copy: high‑pass at 1 Hz + notch at 60 Hz
        raw_hp = raw.copy().filter(l_freq=1.0, h_freq=None, fir_design="firwin")
        raw_hp.notch_filter(60)

        ica = ICA(n_components=30, random_state=42, max_iter="auto")
        ica.fit(raw_hp, picks="eeg")

        # detect ocular and muscle components
        eog_idx, _     = ica.find_bads_eog(raw_hp, ch_name=["HOEG", "VEOG"], measure="correlation", threshold="auto")
        # muscle_idx, _  = ica.find_bads_muscle(raw_hp, threshold=0.5)
        ica.exclude     = list(set(eog_idx))
        if len(ica.exclude) > 0:
            ica.plot_components(picks=ica.exclude)
            print(f"total EOG components: {len(eog_idx)}")
            print(f"idx of EOG components: {eog_idx}")
        # apply the ICA solution in‑place to the high‑pass data
        

        # 2) final filtering: low‑pass at 45 Hz, drop EOG channels
        raw_clean = raw.copy()
        raw_clean = ica.apply(raw_clean)
        # raw_clean.notch_filter(60)
        raw_clean.filter(l_freq=1.0, h_freq=45, fir_design="firwin")
        #.filter(l_freq=0.5, h_freq=45, fir_design="firwin")
        raw_clean.drop_channels(["VEOG", "HOEG"])
        # raw_clean = compute_current_source_density(raw_clean)  # optional CSD

        # 3) events & epoching
        events, _ = mne.events_from_annotations(raw_clean)
        for old, new in mapping.items():
            events[events[:, 2] == old, 2] = new

        epochs = mne.Epochs(
            raw_clean,
            events,
            event_id=new_event_ids,
            tmin=0,
            tmax=60,
            baseline=None,
            preload=True,
        )
        epochs.save(f"{output_dir}/{subject_num}-epochs-epo.fif", overwrite=True)
        subject_num += 1
```
